```python
# Now that we know there are multiple visits we can pull the data from those visits to start analyzing the spectra
spectra = []

for row in fries_rows:
    # I was getting an error something about bytes not str???
    sas_url = row['sas_url'].decode()
    local_path = sas_url.replace(url_root, local_root)

    # From lab04 and others 
    if not os.path.exists(local_path):
        access.remote()
        access.add(row['sas_url']) #line of error?
        access.set_stream()
        access.commit()

    # open the FITS file
    hdul = astropy.io.fits.open(local_path)
    data = hdul[1].data

    wavelength = 10**data['loglam']
    flux = data['flux']
    mjd = row['mjd']

    spectra.append({
        "wavelength": wavelength,
        "flux": flux,
        "mjd": mjd
    })
### Needs to be fixed
```

```python
def line_flux_beta(wl_beta, fl_beta, center, width=40):
    lam_min_beta = center - width
    lam_max_beta = center + width

    mask_beta = (wl_beta > lam_min_beta) & (wl_beta < lam_max_beta)

    # continuum from edges
    cont_mask = ((wl_beta > lam_min_beta) & (wl_beta < lam_min_beta+10)) | \
                ((wl_beta < lam_max_beta) & (wl_beta > lam_max_beta-10))
    cont = np.nanmedian(fl_beta[cont_mask])

    return np.trapz(fl_beta[mask_beta] - cont, wl_beta[mask_beta])

hb_center = 4861 * (1 + z)

hb_times = []
hb_values = []

for s in spectra:
    wl_beta = s["wavelength"]
    fl_beta = s["flux"]
    mjd = s["mjd"]

    f = line_flux_beta(wl_beta, fl_beta, hb_center)
    hb_times.append(mjd)
    hb_values.append(f)
plt.figure(figsize=(8,5))
plt.scatter(hb_times, hb_values)
plt.xlabel("MJD")
plt.ylabel("Hβ Line Flux")
plt.title("Hβ Light Curve")
plt.show();
```
